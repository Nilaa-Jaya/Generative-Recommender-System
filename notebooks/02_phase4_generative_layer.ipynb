```python
import json
from torch.utils.data import Dataset
from transformers import DataCollatorForLanguageModeling, TrainingArguments, Trainer
from peft import get_peft_model, LoraConfig, TaskType
```

```python
# ── 1. Define a lazy tokenizing Dataset ──────────────────────────────────────
class LazyRecDataset(Dataset):
    def __init__(self, path, tokenizer, max_length=512):
        self.path = path
        self.tokenizer = tokenizer
        self.max_length = max_length
        # Precompute byte offsets for each line
        self.offsets = []
        with open(path, "r", encoding="utf-8") as f:
            while True:
                offset = f.tell()
                line = f.readline()
                if not line:
                    break
                self.offsets.append(offset)

    def __len__(self):
        return len(self.offsets)

    def __getitem__(self, idx):
        with open(self.path, "r", encoding="utf-8") as f:
            f.seek(self.offsets[idx])
            obj = json.loads(f.readline())
        prompt = obj["input"] + "\n\n" + obj["output"]
        enc = self.tokenizer(
            prompt,
            padding="max_length",
            truncation=True,
            max_length=self.max_length,
            return_tensors="pt",
        )
        input_ids = enc["input_ids"].squeeze(0)
        return {
            "input_ids": input_ids,
            "attention_mask": enc["attention_mask"].squeeze(0),
            "labels": input_ids.clone(),
        }
```

```python
# ── 2. Instantiate the dataset ────────────────────────────────────────────────
train_dataset = LazyRecDataset("train.jsonl", tokenizer)
```

```python
# ── 3. Wrap model with LoRA (if not already done) ─────────────────────────────
peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.05,
)
model = get_peft_model(model, peft_config)
```

```python
# ── 4. Set up data collator ───────────────────────────────────────────────────
collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)
```

```python
# ── 5. Configure Trainer to use multiple workers ──────────────────────────────
training_args = TrainingArguments(
    output_dir="./genrec_llama2_lora",
    per_device_train_batch_size=1,     # keep or bump to 8 if it fits
    gradient_accumulation_steps=4,     # effective batch = batch_size × accum_steps
    learning_rate=2e-5,
    max_steps=11520,                   # ⇐ limit to ~10 h of training at 0.32 it/s
    fp16=True,
    logging_steps=10,
    save_steps=1000,
    report_to="none",
    dataloader_num_workers=8,
    save_strategy="steps",
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    data_collator=collator,
)
```

```python
# At launch, attempt to resume from the last checkpoint if any:
import os, glob
ckpts = sorted(glob.glob("genrec_llama2_lora/checkpoint-*"), key=os.path.getmtime)
last_ckpt = ckpts[-1] if ckpts else None
```

```python
# ── 6. Launch training ─────────────────────────────────────────────────────────
trainer.train(resume_from_checkpoint=last_ckpt)
```

```python
import time
from itertools import product
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
from google.colab import drive

# ─── load model & tokenizer ───────────────────────────────────────────────────
drive.mount('/content/drive', force_remount=True)
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf", use_fast=True)
tokenizer.pad_token = tokenizer.eos_token
base = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Llama-2-7b-hf",
    torch_dtype=torch.float16,
    device_map="auto",
)
model = PeftModel.from_pretrained(
    base,
    "/content/drive/MyDrive/genrec_llama2_lora",
    torch_dtype=torch.float16,
    device_map="auto",
).eval()

# ─── build prompt ──────────────────────────────────────────────────────────────
few_shot = """
Here are examples of one-sentence recommendations, each using a different hook:

User query: I love character-driven mysteries
Candidate: Gone Girl
Metadata: psychological twist, unreliable narrator, suspense
Recommendation 1: You might enjoy "Gone Girl"—its razor-sharp psychological twists will keep you guessing until the last page.

User query: I want epic world-building
Candidate: Dune
Metadata: desert planet, political intrigue, prophecy
Recommendation 1: You might enjoy "Dune"—its vast desert landscapes and intricate power-struggles will sweep you into a truly alien universe.
""".strip()

query     = "I like sci-fi thrillers"
candidate = "Dune"
metadata  = "sci-fi, epic, political intrigue"

base_prompt = f"""{few_shot}

User query: {query}
Candidate: {candidate}
Metadata: {metadata}

Write three one-sentence recommendations:
Recommendation 1:
Recommendation 2:
Recommendation 3:
"""

# tokenize once
inputs     = tokenizer(base_prompt, return_tensors="pt", truncation=True, max_length=512).to(model.device)
prompt_len = inputs["input_ids"].shape[1]

# ─── helper to run one config ─────────────────────────────────────────────────
def run_config(temp, p, max_toks):
    out = model.generate(
        **inputs,
        do_sample=True,
        temperature=temp,
        top_p=p,
        max_new_tokens=max_toks,
        pad_token_id=tokenizer.eos_token_id,
    )
    gen_text = tokenizer.decode(out[0][prompt_len:], skip_special_tokens=True)
    # extract lines starting with "Recommendation"
    return [line.strip() for line in gen_text.split("\n") if line.strip().startswith("Recommendation")]

# ─── hyper-parameter grid ─────────────────────────────────────────────────────
grid = {
    "temperature":    [0.5, 0.7, 0.9],
    "top_p":          [0.8, 0.9],
    "max_new_tokens": [50, 75],
}

# ─── grid-search loop ─────────────────────────────────────────────────────────
for temp, p, max_toks in product(grid["temperature"],
                                 grid["top_p"],
                                 grid["max_new_tokens"]):
    start = time.time()
    recs  = run_config(temp, p, max_toks)
    dur   = time.time() - start

    print(f"\n=== temp={temp}, top_p={p}, max_new_tokens={max_toks} | {dur:.1f}s ===")
    for r in recs:
        print(r)
```

# C) Integration with Retriever
◻ Dynamic prompt construction

1. Pull in top–K retrieved items & metadata from Phase 3 automatically

2. Loop over them to generate N recommendations in one call

◻ Batch inference

1. Wrap the whole generate procedure into a function that accepts a list of queries

2. Vectorize tokenization + generation for speed

```python
# ─── 1) Suppose `topk` is your list of retrieved items ───────────────────────────
# For example:
topk = [
    {"asin": "B001", "title": "Dune",       "metadata": "sci-fi, epic, political intrigue"},
    {"asin": "B002", "title": "Neuromancer","metadata": "cyberpunk, AI, noir"},
    {"asin": "B003", "title": "Foundation", "metadata": "space opera, mathematics, empire"},
    # … up to your K items
]

query = "I like sci-fi thrillers"

# ─── 2) Build one prompt per item ────────────────────────────────────────────────
def make_prompt(q, candidate, metadata):
    return f"""
You are a recommender bot. Write exactly one sentence:

User query: {q}
Candidate: {candidate}
Metadata: {metadata}
Recommendation:
""".strip()

prompts = [ make_prompt(query, item["title"], item["metadata"])
            for item in topk ]

# ─── 3) Batch‐tokenize all K prompts at once ────────────────────────────────────
batch = tokenizer(
    prompts,
    return_tensors="pt",
    padding=True,
    truncation=True,
    max_length=512
).to(model.device)

prompt_lengths = (batch["attention_mask"].sum(dim=1) - 1).tolist()
# each prompt ends just before the first mask=0

# ─── 4) Generate K recommendations in one call ─────────────────────────────────
outputs = model.generate(
    **batch,
    max_new_tokens=40,
    do_sample=True,
    temperature=0.8,
    top_p=0.9,
    pad_token_id=tokenizer.eos_token_id,
)

# ─── 5) Decode & assemble into a result list ──────────────────────────────────
results = []
for i, out_ids in enumerate(outputs):
    # skip past the prompt tokens
    start = prompt_lengths[i]
    gen_ids = out_ids[start:]
    text    = tokenizer.decode(gen_ids, skip_special_tokens=True).split("\n")[0].strip()
    results.append({
        "asin":           topk[i]["asin"],
        "title":          topk[i]["title"],
        "metadata":       topk[i]["metadata"],
        "recommendation": text
    })

# ─── 6) You now have a list of dicts with your recommendations ──────────────────
for r in results:
    print(f"{r['asin']} – {r['title']} → {r['recommendation']}")
```

```python
import torch
from transformers import StoppingCriteria, StoppingCriteriaList

# 1) stop on the first period
class StopOnPeriod(StoppingCriteria):
    def __call__(self, input_ids, scores, **kwargs):
        return input_ids[0, -1].item() == tokenizer.convert_tokens_to_ids(".")

stop_crit = StoppingCriteriaList([StopOnPeriod()])

def make_recommendation(query, title, tags):
    # pick one hook or fallback to "storytelling"
    hook = tags.split(",")[0].strip() if tags else "storytelling"

    few_shot = """
Here are four examples of exactly one‐sentence recs, each using the TAG field:

User query: I need a laugh‐out‐loud memoir
Candidate: Bossypants
Tag: humor
Recommendation: You might enjoy "Bossypants"—its humor and candid storytelling will have you laughing out loud.

User query: I want sweeping political space opera
Candidate: Foundation
Tag: space opera
Recommendation: You might enjoy "Foundation"—its space opera scope and intricate politics will sweep you to distant galaxies.

User query: I love gothic romance mysteries
Candidate: Rebecca
Tag: mystery
Recommendation: You might enjoy "Rebecca"—its mystery and haunting atmosphere will keep you spellbound.

User query: I like immersive world‐building
Candidate: Dune
Tag: desert planet
Recommendation: You might enjoy "Dune"—its desert planet setting and epic intrigue will transport you to Arrakis.
""".strip()

    prompt = f"""
{few_shot}

User query: {query}
Candidate: {title}
Tag: {hook}
Recommendation: You might enjoy "{title}"—its {hook}""".lstrip()

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=256
    ).to(model.device)

    out = model.generate(
        **inputs,
        max_new_tokens=30,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id,
        stopping_criteria=stop_crit,
    )

    gen = out[0][ inputs["input_ids"].shape[-1] : ]
    text = tokenizer.decode(gen, skip_special_tokens=True).strip()
    return (text.split(".", 1)[0] + ".").strip()

# run it
all_recs = []
for asin, title, tags, _ in topk_info.itertuples(index=False):
    rec = make_recommendation("I like sci-fi thrillers", title, tags)
    all_recs.append({
        "asin":           asin,
        "title":          title,
        "tags":           tags,
        "recommendation": rec
    })

import pprint; pprint.pprint(all_recs)
```
